```python
vocabulary = []
for email in emails_words_list:
    for word in email:
        vocabulary.append(word)

vocabulary = list(set(vocabulary))
len(vocabulary)
```

```python
final_df_dict = {}

final_df_dict['EMAIL'] = []

for email in emails_words_list:
    final_df_dict['EMAIL'].append(email)


for element in vocabulary:
    final_df_dict[element] = []
    for email in emails_words_list:
        count = 0
        for word in email:
            if(word == element):
                count = count+1
        final_df_dict[element].append(count)
```

```python
k_fold_set = []
k_fold_set.append(final_df[:142])
k_fold_set.append(final_df[142:284])
k_fold_set.append(final_df[284:426])
k_fold_set.append(final_df[426:568])
k_fold_set.append(final_df[568:710])
k_fold_set.append(final_df[710:852])
k_fold_set.append(final_df[852:])
```

```python
print(k_fold_set[0].shape)
print(k_fold_set[1].shape)
print(k_fold_set[2].shape)
```

```python
def naive_bayes(k_fold_set, i):
    training_data_list = []
    testing_data = k_fold_set[i]
    for j in range(7):
        if(j != i):
            training_data_list.append(k_fold_set[j])
    training_data = pd.concat(training_data_list)
    
    spam_messages = training_data[training_data['CLASSLABEL'] == 1]
    ham_messages = training_data[training_data['CLASSLABEL'] == 0]
    
    p_spam = len(spam_messages)/len(training_data)
    p_ham = len(ham_messages)/len(training_data)
    

    n_words_per_spam_message = spam_messages['EMAIL'].apply(len)
    n_spam = n_words_per_spam_message.sum()


    n_words_per_ham_message = ham_messages['EMAIL'].apply(len)
    n_ham = n_words_per_ham_message.sum()

 
    n_vocabulary = len(vocabulary)

   
    alpha = 1
    
    parameters_spam = {unique_word:0 for unique_word in vocabulary}
    parameters_ham = {unique_word:0 for unique_word in vocabulary}
    
    for word in vocabulary:
        n_word_given_spam = spam_messages[word].sum() 
        p_word_given_spam = (n_word_given_spam + alpha) / (n_spam + alpha*n_vocabulary)
        parameters_spam[word] = p_word_given_spam
        
        n_word_given_ham = ham_messages[word].sum() 
        p_word_given_ham = (n_word_given_ham + alpha) / (n_ham + alpha*n_vocabulary)
        parameters_ham[word] = p_word_given_ham
        
    
    correct_classi = 0
    incorrect_classi = 0
    for k in range(len(testing_data)):
        p_spam_given_message = p_spam
        p_ham_given_message = p_ham
        word_list_testing = list(testing_data.iloc[k, 1])
        for word in word_list_testing:
            if word in parameters_spam:
                p_spam_given_message *= parameters_spam[word]
            if word in parameters_ham: 
                p_ham_given_message *= parameters_ham[word]
                
        if p_ham_given_message >= p_spam_given_message:
            if(testing_data.iloc[k, 0] == 0):
                correct_classi = correct_classi+1
            else:
                incorrect_classi = incorrect_classi+1
            
        elif p_ham_given_message < p_spam_given_message:
            if(testing_data.iloc[k, 0] == 1):
                correct_classi = correct_classi+1
            else:
                incorrect_classi = incorrect_classi+1
        
    return ((correct_classi) / (incorrect_classi + correct_classi))
    
    
```

```python
accuracy_vals = []
for i in range(7):
    accuracy_vals.append(naive_bayes(k_fold_set, i))

```

```python
for i in range(len(accuracy_vals)):
    print("Fold: " + str(i+1) + ", " + str(accuracy_vals[i]))
```

```python
avg_accuracy = (sum(accuracy_vals))/(len(accuracy_vals))
print(avg_accuracy)
```
